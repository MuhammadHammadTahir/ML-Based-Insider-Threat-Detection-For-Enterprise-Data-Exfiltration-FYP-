```python
import os
import pandas as pd
import json
```

```python
INPUT_FOLDER = r"C:\Users\employee\Desktop\genrator\input"
OUTPUT_FOLDER = r"C:\Users\employee\Desktop\genrator\output"
ROWS_TO_KEEP = 150
```

```python
for filename in os.listdir(INPUT_FOLDER):
    if filename.endswith("_new.csv"):
        input_path = os.path.join(INPUT_FOLDER, filename)
        
        print(f"Processing: {filename}")
        
        # Read only first N rows
        df = pd.read_csv(input_path, nrows=ROWS_TO_KEEP)

        # Clean ID
        df['id'] = df['id'].str.replace(r'[{}]', '', regex=True)
        
        # Rename date field
        df.rename(columns={"date": "@timestamp"}, inplace=True)

        # Convert to ISO format
        df["@timestamp"] = pd.to_datetime(df["@timestamp"], dayfirst=True).dt.strftime('%Y-%m-%dT%H:%M:%S')

        # Fix NaN → null
        df = df.astype(object).where(pd.notnull(df), None)


        # Create new filename (overwrite old if exists)
        base_name = os.path.splitext(filename)[0]
        new_filename = f"{base_name}.json"
        output_path = os.path.join(OUTPUT_FOLDER, new_filename)
        
        # Convert each row to JSON and overwrite file
        with open(output_path, "w", encoding="utf-8") as f:
           for record in df.to_dict(orient="records"):
              f.write(json.dumps(record) + "\n")
        
        print(f"Overwritten with JSON logs: {new_filename}")

print("Done ✅")
```

```python
for filename in os.listdir(INPUT_FOLDER):
    if filename.endswith(".csv"):
        input_path = os.path.join(INPUT_FOLDER, filename)
        
        print(f"Processing: {filename}")
        
        # Read only first 20 rows (FAST - does not load full file)
        df = pd.read_csv(input_path, nrows=ROWS_TO_KEEP)
        
        # Create new filename
        base_name = os.path.splitext(filename)[0]
        new_filename = f"{base_name}_new.csv"
        output_path = os.path.join(OUTPUT_FOLDER, new_filename)
        
        # Save to new file
        df.to_csv(output_path, index=False)
        
        print(f"Saved: {new_filename}")

print("Done ✅")
```
